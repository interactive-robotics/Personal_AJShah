```python
import networkx as nx
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [35,20]
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
green = '#238960'
red = '#701e28'
blue = '#1e1f70'



G.add_node('Fork', color = red)
G.add_node('Spoon',color = red)
G.add_node('Plate', color = red)
G.add_node('Small plate', color = red)
G.add_node('Bowl', color = red )
G.add_node('Knife', color = red)
G.add_node('Mug',color = red)
G.add_node('Glass', color = red)
G.add_node('Start', color = blue)


G.add_edge('Plate', 'Small plate')
G.add_edge('Small plate', 'Bowl')
#G.add_edge('Plate', 'Bowl')

for node in ['Plate','Fork','Spoon','Knife','Mug','Glass']:
    G.add_edge('Start', node)
```

```python
def Render(G=G):
    node_colors = [node[1] for node in G.nodes('color')]
    pos = nx.drawing.nx_agraph.graphviz_layout(G, prog='dot')
    nx.draw_networkx(G, pos, with_labels=True, node_size = 20000, node_color = node_colors, font_size = 32, font_color='w', width = 2.5, arrowsize=40)
    plt.box(False)

def save(name):
    f = plt.gcf()
    f.savefig(name, bbox_inches = 'tight', transparent = True)

    
Render()
save('Initial.png')
```

```python
colors = {'Start':{'color':green}}
for node in ['Fork','Spoon','Plate','Knife','Mug','Glass']:
    colors[node] = {'color':blue}
nx.set_node_attributes(G,colors)
Render()
save('start.png')

```

```python
def setColor(node, color):
    colors = {node: {'color': color}}
    nx.set_node_attributes(G,colors)
    return colors
setColor('Mug',green)
Render()
save('mug.png')
```
